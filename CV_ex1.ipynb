```python
#this function is going to detect lines in the edge image using the algorithm we saw in lecture:
#loop on edge points and on theta, and increment in the accumlator accordingly and it return the accumlator
def hough_line(edge_image):
    # theta array in range (-90,90) as we defined in lecture
    thetas = np.deg2rad(np.arange(-90.0, 90.0))
    # get the shape of the image
    width, height = edge_image.shape
    #We get the maximum lenght a diagonal can be using the image size 
    diagonal_length = np.ceil(np.sqrt(width * width + height * height))
    #This line generates an array of possible distances of the lines from the origin.
    # The distance ranges from -diagonal_length to diagonal_length.   
    possible_distances = np.linspace(-diagonal_length, diagonal_length, int(diagonal_length * 2.0))

    #we define arrays for the value of cos and sin for each theta value from thetas.
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)
    num_thetas = len(thetas)

    #define our accumulator 2d matrix of size of the diagonal lenght and num of thetas
    accumulator = np.zeros((2 * int(diagonal_length), num_thetas), dtype=np.uint64)
   
    # Extract points that were detected as an edge in edge_image
    edge_points = np.argwhere(edge_image != 0)

    # Vote in the hough accumulator
    for y, x in edge_points:
        for theta in range(num_thetas):
            # Calculate distance. diagonal_length is added for a positive index
            #the reason for that is that the numpy doesnt accept negative values so we shifted
            #the values to be postive.
            distance = round(x * cos_t[theta] + y * sin_t[theta]) + diagonal_length
            accumulator[int(distance), theta] += 1
    return accumulator, thetas, possible_distances
```

```python
# def find_lines(accumulator, threshold):
#     # Initialize an empty list to hold the peaks
#     peaks = []

#     # Get the height and width of the accumulator
#     height, width = accumulator.shape

#     # Iterate over the entire accumulator
#     for i in range(height):
#         for j in range(width):
#             # If the value at (i, j) is greater than the threshold, it's a peak
#             if accumulator[i, j] >= threshold:
#                 peaks.append((i, j))

#     return peaks

from scipy.ndimage.filters import maximum_filter

def find_lines(accumulator, threshold, size=100):
    # Apply a maximum filter to find local maxima
    filtered = maximum_filter(accumulator, size=size)

    # The peaks are the coordinates where the original accumulator is equal to the filtered result
    peaks = [(i, j) for i in range(accumulator.shape[0]) for j in range(accumulator.shape[1]) 
             if accumulator[i, j] == filtered[i, j] and accumulator[i, j] >= threshold]

    return peaks
```

```python
def draw_lines(img, lines):
    line_count = 0
    
    for rho, theta in lines:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        scale = img.shape[0] + img.shape[1]

        x1 = int(x0 + scale * (-b))
        y1 = int(y0 + scale * (a))
        x2 = int(x0 - scale * (-b))
        y2 = int(y0 - scale * (a))
        cv2.line(img, (x1, y1), (x2, y2), (255, 255, 0), 2)
        
        line_count += 1

    return img, line_count
```

```python
def plot_peaks(accumulator, peaks, neighborhood_size=10):
    plt.figure(figsize=(10, 10))
    plt.imshow(accumulator, cmap='gray', aspect='auto')

    for peak in peaks:
        y, x = peak
        plt.plot([x-neighborhood_size/2, x-neighborhood_size/2, x+neighborhood_size/2, x+neighborhood_size/2, x-neighborhood_size/2],
                 [y-neighborhood_size/2, y+neighborhood_size/2, y+neighborhood_size/2, y-neighborhood_size/2, y-neighborhood_size/2],
                 'w-')

    plt.title('Hough transform with peaks')
    plt.show()
```

```python
import itertools

def find_intersections(lines, image_width, image_height):
    intersections = []
    for pair in itertools.combinations(lines, 2):
        rho1, theta1 = pair[0]
        rho2, theta2 = pair[1]

        # Check if the lines are parallel
        if np.isclose(theta1, theta2):
            continue

        # Calculate the intersection of the lines
        try:
            a = np.array([[np.cos(theta1), np.sin(theta1)],
                          [np.cos(theta2), np.sin(theta2)]])
            b = np.array([rho1, rho2])
            x, y = np.linalg.solve(a, b)

            # Check if the intersection is within the image frame
            if 0 <= x < image_width and 0 <= y < image_height:
                intersections.append((x, y))
        except np.linalg.LinAlgError:
            continue  # Skip this pair of lines if the matrix is singular

    return intersections
```

```python
from itertools import combinations
def find_triangles(intersections):
    triangles = []
    for (P1, P2, P3) in combinations(intersections, 3):
        area = 0.5 * abs(P1[0] * (P2[1] - P3[1]) + P2[0] * (P3[1] - P1[1]) + P3[0] * (P1[1] - P2[1]))
        if area > 1e-6:  # Check for non-collinearity
            triangles.append((P1, P2, P3))
    return triangles
```

```python


def sliding_window_hough_transform(image, window_size, step_size_right, step_size_bottom):
    # Get the dimensions of the image
    image_height, image_width = image.shape[:2]

    # Initialize a list to store the results
    results = []

    # Slide the window over the image
    for y in range(0, image_height, step_size_bottom):
        for x in range(0, image_width, step_size_right):
            # Extract the window from the image
            window = image[y:y + window_size, x:x + window_size]

            # Apply the Hough transform on the window
            edge_map = cv2.Canny(window, 100, 200)
            edge_map = cv2.normalize(edge_map, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
            accumulator, thetas, possible_distances = hough_line(edge_map)

            # Define the threshold for a peak
            threshold = 40

            # Use the find_lines function to find the peaks in the accumulator
            peaks = find_lines(accumulator, threshold)

            # Each row in peaks is the (y, x) coordinates of a peak in the accumulator
            # To find the corresponding (rho, theta) parameters for each peak, use the rhos and thetas arrays
            lines = [(possible_distances[i], thetas[j]) for i, j in peaks]
            
            lines =non_maximum_suppression(lines)
            #adjusted_lines=lines
            
            adjusted_lines = []
            for rho, theta in lines:
                adjusted_rho = rho + x * np.cos(theta) + y * np.sin(theta)
                adjusted_lines.append((adjusted_rho, theta))


            copy_img=window.copy()
            draw_lines(copy_img,adjusted_lines)
            
            # Detect triangles in the window
            intersections = find_intersections(adjusted_lines, window_size, window_size)
            print(intersections)
            print(x,y)
            #! possible_triangles = find_triangles(intersections)
            
            draw_intersections(copy_img, intersections, x, y)

            plt.imshow(copy_img)
            plt.title('detected lines')
            plt.show()

            

            # Store the results
            #! results.append((x, y, lines, possible_triangles, intersections))
            results.append((x, y, adjusted_lines , intersections))

    return results
```
