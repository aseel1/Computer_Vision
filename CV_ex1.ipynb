```python
#this function is going to detect lines in the edge image using the algorithm we saw in lecture:
#loop on edge points and on theta, and increment in the accumlator accordingly and it return the accumlator
def hough_line(edge_image):
    # theta array in range (-90,90) as we defined in lecture
    thetas = np.deg2rad(np.arange(-90.0, 90.0))
    # get the shape of the image
    width, height = edge_image.shape
    #We get the maximum lenght a diagonal can be using the image size 
    diagonal_length = np.ceil(np.sqrt(width * width + height * height))
    #This line generates an array of possible distances of the lines from the origin.
    # The distance ranges from -diagonal_length to diagonal_length.   
    possible_distances = np.linspace(-diagonal_length, diagonal_length, int(diagonal_length * 2.0))

    #we define arrays for the value of cos and sin for each theta value from thetas.
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)
    num_thetas = len(thetas)

    #define our accumulator 2d matrix of size of the diagonal lenght and num of thetas
    accumulator = np.zeros((2 * int(diagonal_length), num_thetas), dtype=np.uint64)
   
    # Extract points that were detected as an edge in edge_image
    edge_points = np.argwhere(edge_image != 0)

    # Vote in the hough accumulator
    for y, x in edge_points:
        for theta in range(num_thetas):
            # Calculate distance. diagonal_length is added for a positive index
            #the reason for that is that the numpy doesnt accept negative values so we shifted
            #the values to be postive.
            distance = round(x * cos_t[theta] + y * sin_t[theta]) + diagonal_length
            accumulator[int(distance), theta] += 1
    return accumulator, thetas, possible_distances
```

```python
def draw_lines(img, lines):
    line_count = 0
    
    for rho, theta in lines:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        scale = img.shape[0] + img.shape[1]

        x1 = int(x0 + scale * (-b))
        y1 = int(y0 + scale * (a))
        x2 = int(x0 - scale * (-b))
        y2 = int(y0 - scale * (a))
        cv2.line(img, (x1, y1), (x2, y2), (0, 0, 0), 2)        
        line_count += 1

    return img, line_count
```

```python
import itertools

def find_intersections(lines, image_width, image_height):
    intersections = []
    for pair in itertools.combinations(lines, 2):
        rho1, theta1 = pair[0]
        rho2, theta2 = pair[1]

        # Check if the lines are parallel
        if np.isclose(theta1, theta2):
            continue

        # Calculate the intersection of the lines
        try:
            a = np.array([[np.cos(theta1), np.sin(theta1)],
                          [np.cos(theta2), np.sin(theta2)]])
            b = np.array([rho1, rho2])
            x, y = np.linalg.solve(a, b)

            # Check if the intersection is within the image frame
            if 0 <= x < image_width and 0 <= y < image_height:
                intersections.append((x, y))
        except np.linalg.LinAlgError:
            continue  # Skip this pair of lines if the matrix is singular

    return intersections
```

```python

def filter_intersections(intersections, edge_map):
    # Create a blank mask with the same size as the edge map
    mask = np.zeros_like(edge_map)
    
    # Draw the intersection points on the mask
    for point in intersections:
        cv2.circle(mask, (int(point[0]), int(point[1])), 1, (255, 255, 255), -1)
    
    # Perform a bitwise_and operation with the edge map
    result = cv2.bitwise_and(mask, edge_map)
    
    # Filter the intersections
    filtered_intersections = [point for point in intersections if result[int(point[1]), int(point[0])] != 0]
    
    return filtered_intersections
```

```python
def line_intersection(line1, line2):
    rho1, theta1 = line1
    rho2, theta2 = line2

    # Check if the lines are parallel
    if np.isclose(theta1, theta2):
        return None

    # Calculate the intersection of the lines
    try:
        a = np.array([[np.cos(theta1), np.sin(theta1)],
                      [np.cos(theta2), np.sin(theta2)]])
        b = np.array([rho1, rho2])
        x, y = np.linalg.solve(a, b)

        return (x, y)
    except np.linalg.LinAlgError:
        return None  # Return None if the matrix is singular
```

```python
def calculate_angle(P1, P2, P3):
    a = np.array(P1) - np.array(P2)
    b = np.array(P3) - np.array(P2)
    cosine_angle = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    angle = np.arccos(np.clip(cosine_angle, -1.0, 1.0))
    return angle
    
def classify_triangle(P1, P2, P3, d1, d2, d3, length_threshold=10, angle_threshold=0.3):
    sides = sorted([d1, d2, d3])
    angles = sorted([calculate_angle(P1, P2, P3), calculate_angle(P2, P3, P1), calculate_angle(P3, P1, P2)])

    if np.isclose(angles[2], np.pi / 2, atol=angle_threshold) and all(angle < np.pi / 2 for angle in angles[:2]):
        return 'right-angled'
    elif np.isclose(sides[0], sides[1], atol=length_threshold) and np.isclose(sides[1], sides[2], atol=length_threshold):
        return 'equilateral'
    elif (np.isclose(sides[0], sides[1], atol=length_threshold) or
          np.isclose(sides[1], sides[2], atol=length_threshold) or
          np.isclose(sides[0], sides[2], atol=length_threshold)):
        return 'isosceles'
    else:
        return 'scalene'
```

```python
def remove_close_lines(lines, threshold):
    # Sort the lines by their starting x-coordinate
    lines.sort(key=lambda line: line[0])

    # Initialize the list of unique lines with the first line
    unique_lines = [lines[0]]

    for line in lines[1:]:
        # Check the distance to the last unique line
        if abs(line[0] - unique_lines[-1][0]) > threshold:
            # If the distance is greater than the threshold, add the line to the list of unique lines
            unique_lines.append(line)

    return unique_lines
```
